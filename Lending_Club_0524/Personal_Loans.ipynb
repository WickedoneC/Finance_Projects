```python
import pandas as pd
import numpy as np
import seaborn as sns

def prettyTable(df, col):
    a = df[col].value_counts(dropna = False)
    b = df[col].value_counts(normalize = True, dropna = False)
    c = pd.concat([a, b], axis = 1)
    c.columns = [col + '_count', col + '_ratio']
    return c
```

```python
# term - drop 'months' and convert to int
# emp_title - drop column
# emp_length = remove + years and convert to int
# issue_d - convert to date
    # then create new columns with months since date
    
# title = dropm column
# zip_code = drop column

# earliest_cr_line - convert to date
    # then create new columns with months since date
# last_pymnt_d - convert to date
    # then create new columns with months since date
# last_credit_pull_d - convert to date
    # then create new columns with months since date
    
# loan status - fully paid + current vs the remaining categories
```

```python
df.head()
```

```python
prettyTable(df, 'term')
df['term'] = df['term'].replace({' months':''}, regex = True)
prettyTable(df, 'term')
```

```python
df.term = df.term.astype(int)
```

```python
# display(len(df.emp_title.unique()))
df = df.drop('emp_title', axis = 1)
```

```python
# display(prettyTable(df, 'emp_length'))

# rounding <1 year to 1 year
df['emp_length'] = df['emp_length'].map(lambda x: x.lstrip('<').rstrip('+ years'))

prettyTable(df, 'emp_length')
```

```python
df.emp_length = df.emp_length.astype(int)
```

```python
# converting dates and finding time since issue in months
df['issue_d'] = pd.to_datetime(df['issue_d'], format = '%b-%Y')

df['current_d'] = 'May-2024'
df['current_d'] = pd.to_datetime(df['current_d'])


df['months_since_issue'] = (df.current_d-df.issue_d).astype('timedelta64[M]') + 2
```

```python
# drop title
df = df.drop('term', axis = 1)
```

```python
# drop zip code
df = df.drop('zip_code', axis = 1)
```

```python
# converting dates and finding time since issue in months
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], format = '%b-%Y')

df['months_since_earliest_cr_line'] = (df.current_d-df.earliest_cr_line).astype('timedelta64[M]') + 2



df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], format = '%b-%Y')

df['months_since_last_pymnt_d'] = (df.current_d-df.last_pymnt_d).astype('timedelta64[M]') + 2



df['last_credit_pull_d'] = pd.to_datetime(df['last_credit_pull_d'], format = '%b-%Y')

df['months_since_last_pymnt_d'] = (df.current_d-df.last_credit_pull_d).astype('timedelta64[M]') + 2

```

```python
# drop issue_d
# drop earliest_cr_line
# drop last_pymnt_d
# drop last_credit_pull_d

drops = ['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d']

df = df.drop(drops, axis = 1)
```

```python
# loan status - fully paid + current vs the remaining categories
prettyTable(df, 'loan_status')
```

```python
df['loan_outcomes'] = loan_outcomes
prettyTable(df, 'loan_outcomes')
```
